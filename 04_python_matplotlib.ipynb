### Figure

```python
import matplotlib.pyplot as plt
fig = plt.figure()                      # creates a new Figure object
fig = plt.figure(figsize=(8, 6))        # creates a new Figure object with a size
                                        # of 8x6 inches (width x height)
fig = plt.figure(facecolor='lightblue') # new figure with a light blue background
```

### Axes

```python
import matplotlib.pyplot as plt
fig = plt.figure()

ax = fig.add_subplot(2,2,1) # creates a new Axes object in a 2x2 grid
                            # and returns the first Axes object (top left)

ax = fig.add_subplot(221)   # identical to the line above
ax = fig.add_subplot(224)   # creates the fourth Axes object (bottom right)

#ax = fig.add_subplot(22)    # ax is now a 2x2 array of axes; the cells are
#                            # addressed with ax[0,0], ax[0,1], ax[1,0] and ax[1,1]
```

### First Plot

```python
import matplotlib.pyplot as plt
import numpy as np
l = plt.plot(np.random.randn(100).cumsum())

line = l[0]        # access to 1st line obj.
ax   = l[0].axes   # access to Axes
fig  = l[0].figure # access to Figure

#Alternative:
ax  = plt.gca() # access to current Axes
fig = plt.gcf() # access to current Figure
```

### Colors, Markers and Lines

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2*np.pi, 20)
x = np.cos(t)
y = np.sin(t)

plt.plot(x, y, color='red',
         marker='o',
         linestyle='dashed')

# change color to blue afterwards
plt.gca().lines[0].set_color('blue')
```

### Ticks and Labels

```python
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(5, 7))
ax=fig.add_subplot(111)
ax.plot(np.random.randn(100).cumsum())

ticks = ax.set_xticks([0, 50, 100])
labels = ax.set_xticklabels(['one', 'two', 'three'],
                    rotation=30, fontsize='small')

ax.set_title('Important data')
ax.set_xlabel('Samples')
ax.set_ylabel('Cumulative sum')
```

### Legends

```python
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(np.random.randn(100).cumsum(), 'C0', label='one')
ax.plot(np.random.randn(100).cumsum(), 'C1', label='two')
ax.plot(np.random.randn(100).cumsum(), 'C2', label='three')

ax.legend(loc='best', frameon=True, title='Legend')
```

### Saving Figures

```python
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(np.random.randn(100).cumsum())

plt.savefig('diagramm.png', dpi=400)
```

### Plotting with Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
ts = pd.Series(np.random.randn(180),
    index=pd.date_range('1/1/2025', periods=180))
ts = ts.cumsum()

plt.figure(figsize=(4, 5))
ts.plot()
plt.xticks(rotation=45)
```

### Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/ktnspr/py4ac/main/automobile.csv')[['make', 'horsepower']]
makers = ['audi', 'bmw', 'chevrolet', 'dodge', 'ford', 'jaguar', 'mercedes-benz',
          'porsche']
df = df[df['make'].isin(makers)]
plt.figure(figsize=(10, 2))
sns.barplot(x='horsepower', y='make', data=df, orient='h', errorbar='sd', width=0.7)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Daten einlesen
df = pd.read_csv('https://raw.githubusercontent.com/ktnspr/py4ac/main/automobile.csv')[['make', 'num-of-doors']]
df = df.dropna()
df['cyl'] = df['num-of-doors'].astype(int)
makers = ['audi', 'bmw', 'chevrolet', 'dodge', 'ford', 'jaguar',  'porsche']
df = df[df['make'].isin(makers)]
plt.figure(figsize=(10, 2))
sns.countplot(data=df, x='make', hue='num-of-doors', width=0.8, palette='Set2')
```
